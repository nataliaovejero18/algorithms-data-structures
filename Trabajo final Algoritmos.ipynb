```python
class Cliente:
    def __init__(self, nombre, id):
        self.nombre = nombre
        self.id = id
        self.pedidos = []

    def agregar_pedido(self, cantidad):
        if cantidad > 0:
            self.pedidos.append(cantidad)
        else:
            print("Monto inválido, no se agregó el pedido.")

    def promedio_pedidos(self):
        if not self.pedidos:
            return 0
        return sum(self.pedidos) / len(self.pedidos)

    def mayor_pedido(self):
        if not self.pedidos:
            return None
        return max(self.pedidos)
        
def cliente_promedio_mayor(lista_clientes):
    return max(lista_clientes, key=lambda c: c.promedio_pedidos())

# Clientes
c1 = Cliente("Natalia", 118)
c2 = Cliente("Hugo", 119)
c3 = Cliente("Iván", 120)

# Pedidos
c1.agregar_pedido(100)
c1.agregar_pedido(600)

c2.agregar_pedido(300)
c2.agregar_pedido(200)

c3.agregar_pedido(1000)

# Promedios
print("Promedios:")
print(c1.nombre, c1.promedio_pedidos())
print(c2.nombre, c2.promedio_pedidos())
print(c3.nombre, c3.promedio_pedidos())

# Cliente con mayor promedio
mejor = cliente_promedio_mayor([c1, c2, c3])
print(f"El cliente con mayor promedio es: {mejor.nombre}")

```

#### En un videojuego de aventuras, el personaje principal atraviesa diferentes niveles sobre los cuales va cazando super poderes representados por objetos que va guardando en una mochila, sin embargo este tiene una gran restricción solo puede acceder al último objeto que guardó, y si lo saca, el anterior queda disponible.

#### Cada objeto tiene un nombre, un peso y un tipo (arma, poción, herramienta, etc.).

#### Requerimientos: 

#### 1.	Implementa un TAD Mochila con las operaciones:

 *  guardar(objeto) → Guarda un nuevo objeto en la parte superior de la mochila.
 *  usar() → Usa y elimina el último objeto guardado.
 *  ver_ultimo() → Devuelve el nombre del último objeto sin quitarlo.
 *  vacía() → Verifica si la mochila está vacía.

#### 2.	¿Qué TAD utilizaste para dar solución al ejercicio? ¿por qué?

```python
class Objeto:
    def __init__(self, nombre, peso, tipo):
        self.nombre = nombre
        self.peso = peso
        self.tipo = tipo

    def __str__(self):
        return f"{self.nombre} ({self.tipo}, {self.peso}kg)"
        
class Mochila:
    def __init__(self):
        self.pila = []

    def guardar(self, objeto):
        self.pila.append(objeto)

    def usar(self):
        if self.vacia():
            return "Mochila vacía."
        return self.pila.pop()

    def ver_ultimo(self):
        if self.vacia():
            return "Mochila vacía."
        return self.pila[-1]

    def vacia(self):
        return len(self.pila) == 0

# Crea rlos Objetos
pala = Objeto("Pala acero", 5, "herramienta")
escudo = Objeto("Escudo titanio", 1, "arma")
arco = Objeto("Arco legendario", 4, "arma")

# Crear la mochila
mochila = Mochila()

# Guardar objetos
mochila.guardar(pala)
mochila.guardar(escudo)
mochila.guardar(arco)

# Ver último objeto
print("Último objeto:", mochila.ver_ultimo())

# Usar el último objeto
print("Utilizando:", mochila.usar())

# Ver el último después de usar
print("Último objeto después de usarlo:", mochila.ver_ultimo())

```

## EJERICICO 3 (Análisis de Calificaciones)

#### Supón que tienes un diccionario donde las llaves son nombres de estudiantes y los valores son listas con sus calificaciones.

#### Crea diferentes funciones para realizar lo siguiente: 
#### 1.	**Promedio por estudiante**
Documentación: https://docs.python.org/3/library/functions.html
Crea una función que reciba el diccionario y retorne otro diccionario con el nombre del estudiante y su promedio (redondeado a dos cifras decimales).
#### 2.	**Estudiante con mejor promedio**
Documentación: https://docs.python.org/3/library/functions.html#max
Devuelve el nombre del estudiante con el promedio más alto.

#### 3.	**Reemplazo de notas bajas**
Documentación: https://docs.python.org/es/3.13/tutorial/datastructures.html
Crea una función que reciba el diccionario y devuelva una nueva estructura donde todas las notas menores a 3.0 hayan sido reemplazadas por 3.0. (No modifiques el original.)

#### 4.	**Estudiantes sobresalientes**
Crea una función que devuelva un set con los nombres de los estudiantes que tuvieron todas sus calificaciones mayores o iguales a 4.0.


```python
def promedio_estudiantes(diccionario):
    promedios = {}
    for estudiante, notas in diccionario.items():
        promedio = round(sum(notas) / len(notas), 2)
        promedios[estudiante] = promedio
    return promedios

def mejor_promedio_estudiante(diccionario):
    promedios = promedio_estudiantes(diccionario)
    return max(promedios, key=promedios.get)

def reemplazar_notasbajas(diccionario):
    nuevo_diccionario = {}
    for estudiante, notas in diccionario.items():
        nuevas_notas = [max(n, 3.0) for n in notas]
        nuevo_diccionario[estudiante] = nuevas_notas
    return nuevo_diccionario

def estudiantes_sobresalientes(diccionario):
    sobresalientes = {nombre for nombre, notas in diccionario.items() if all(n >= 4.0 for n in notas)}
    return sobresalientes

# Prueba
calificaciones = {
    "Hugo": [2.0, 3.8, 4.2],
    "Iván": [5.9, 6.0, 5.4],
    "Cristina": [7.8, 8.6, 8.0],
    "Daniel": [6.0, 8.5, 9.1]
}

print("Promedio de calificaciones:")
print(promedio_estudiantes(calificaciones))

print("Estudiante con mejor promedio de calificaiones:")
print(mejor_promedio_estudiante(calificaciones))

print("Lista con notas corregidas:")
print(reemplazar_notasbajas(calificaciones))

print("Estudiantes con calificaciones sobresalientes:")
print(estudiantes_sobresalientes(calificaciones))
```

* def promedio_estudiantes: utilizo *sum(notas)* para sumar todas las notas, luego dividirlas para sacer el promedio *len(notas)* y por úlitmo *round(valor,2)* para redondéar el resultado
* def mejor_promedio_estudiante: calculo los promedios primero y luego encotrar el estudinte con valor más alto *max()*
* def reemplazar_notasbajas: reemplazo las notas bajas a 3.0 *max(n,3.0)* y luego se crea un nuevo diccionario para las nuevas notas
* def estudiantes_sobresalientes: compruebo que las notas sean más alta que un 4.0 *all()*

## EJERICICO 4 (Clasificación de Clientes en una Tienda Online)

#### Una tienda online quiere mejorar la atención y las promociones que ofrece a sus clientes, clasificándolos en tres categorías:
   *	Clientes Premium
   *	Clientes Frecuentes
   *	Clientes Ocasionales
#### Para ello, ha definido el siguiente conjunto de reglas basadas en el comportamiento de compra:
#### **Reglas del Árbol de Decisión:**
#### 1.	Si el cliente ha realizado más de 10 compras, entonces:
 *  Si ha gastado más de 10,000 euros en total, es un Cliente Premium.
 * Si no, es un Cliente Frecuente.
#### 2. Si el cliente ha hecho entre 5 y 10 compras, entonces:
 * Si ha dejado al menos 1 reseña positiva, es un Cliente Frecuente.
 * Si no, es un Cliente Ocasional.
#### 3.	Si el cliente ha hecho menos de 5 compras, es un Cliente Ocasional.

#### Implementa un árbol de decisión en Python que, dadas las características de un cliente (número de compras, total gastado y si dejó o no una reseña positiva), determine su categoría.


```python
def clasificación_cliente(cliente):
    compras = cliente["compras"]
    dinero  = cliente["dinero_gastado"]
    reseña  = cliente["resena_positiva"]

    if compras > 10:
        if dinero > 10000:
            return "Cliente Premium"
        else:
            return "Cliente Frecuente"
    elif 5 <= compras <= 10:
        if reseña:
            return "Cliente Frecuente"
        else:
            return "Cliente Ocasional"
    else:
        return "Cliente Ocasional"
    

# Crear perfil de clientes
clientes = [
    {"nombre": "Cristina", "compras": 18, "dinero_gastado": 14000, "resena_positiva": False},  
    {"nombre": "Iván", "compras": 11, "dinero_gastado": 6000, "resena_positiva": True},   
    {"nombre": "Hugo", "compras": 6, "dinero_gastado": 7500, "resena_positiva": False},    
    {"nombre": "Marta", "compras": 5, "dinero_gastado": 4000, "resena_positiva": True},    
    {"nombre": "Daniel", "compras": 2, "dinero_gastado": 1000, "resena_positiva": True},    
]

for cliente in clientes:
    print (f"{cliente['nombre']}: {clasificación_cliente(cliente)}")

```
